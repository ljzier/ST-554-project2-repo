## Part I - Creating a Class

I've created a data folder on my github repository and stored the air quality data there.  This part of the project utilizes the air quality data set available at the UCI Machine Learning Repository (https://archive.ics.uci.edu/dataset/360/air+quality). This data is time series data (data recorded over time) for air quality measurements of pollutants in an Italian city. The dataset consists of 9,358 hourly-averaged observations collected from five metal oxide chemical sensors integrated into an air quality multisensor device. The data were gathered over a one-year period from March 2004 through February 2005 (De Vito, 2008).

The pollutants measured, which are "true" (gold standard) measurements, are the following:

- CO(GT) (CO concentration)
- C6H6(GT) (Benzene concentration)
- NMHC(GT) (Non Metanic HydroCarbon concentration)
- NOx(GT) (NOx concentration)
- NO2(GT) (NO2 concentration)
- T (temperature)
- RH (Relative humidity)
- AH (Absolute Humidity)

I've included this information here because it is important to understand the data you are analyzing, though the focus of this project is to program in Spark SQL, pandas-on-Spark, and pandas. Let's get to it.

```python
# imports 
from pyspark.sql import DataFrame
from pyspark.sql import functions as F
from functools import reduce
from pyspark.sql.types import *
import pandas as pd

from pyspark.sql import SparkSession


spark = SparkSession.builder.appName("Project2").config("spark.sql.ansi.enabled", "false").getOrCreate()
```

```python
# imports to run when the kernel gets restarted

import Spark_class
from Spark_class import SparkDataCheck

import importlib
importlib.reload(Spark_class)
```

    After loading the data, I checked the numeric data types for missing values indicated by an entry of -200. I replaced those with None to give NULL values. I also dropped the first column of indexes.  I don't need it and it gives me warnings later because there is no heading to the column.

```python

# Loading the air data using the from_csv method 
air = Spark_class.SparkDataCheck.from_csv(spark, "data/air.csv")
   
#air.df.printSchema(5)

#I see this first column is just index values so I drop it to prevent warnings
air.df=air.df.drop("_c0")

#get numeric columns
numeric_types = ['int', 'bigint', 'float', 'double', 'decimal', 'long', 'short']

#cleaning up : replacing the -200 values with NULL
for c, dtype in air.df.dtypes:
    if dtype in numeric_types:
        air.df=air.df.withColumn(c, F.when(F.col(f'`{c}`')== -200, None).otherwise(F.col(f'`{c}`')))
    
air.df.show(5)
```

    In order to test the string methods, I needed to add a few more columns with string variables.  It made sense to quantify temperature and relative humidity. I also needed to ensure the NULL values didn't end up in one of the other categories to falsely record the level.

```python
# making some string columns that hopefully make sense

air.df = air.df.withColumn('T_level',F.when(F.col('T') > 20, 'high').
                           when(F.col('T') > 10, 'medium').when(F.col('T') >= 0, 'low').otherwise(None))

air.df = air.df.withColumn('RH_level',F.when(F.col('RH') > 60,'high').
                           when(F.col('RH')>30, 'medium').when(F.col('RH') >= 0, 'low').otherwise(None))

air.df.show()

#checking to make sure the NULL values don't end up in one of the other categories to falsely record the level
result = air.check_null("RH")
result.df.filter(F.col('null_check') == True).show(5)
```

## Part II - Spark SQL and pandas-on-Spark APIs

In this section I'm going to do some basic analysis using two different Spark APIs (Application Programming Interfaces): pandas-on-Spark and Spark SQL. Both allow us to work with large datasets using Spark as the background engine, but pandas-on-Spark mimics the pandas syntax while Spark SQL uses SQL-style commands.  I will perform some analysis on using each of these and note differences along the way.

The data is weekly NFL data. I will be focusing on quarterback stats from the 2005-2023 regular seasons.

### pandas-on-Spark

In this section I use pandas-on-Spark commands to look at the NFL data.

```python
# Spark and pandas imports
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
import pyspark.pandas as ps

ps.set_option('compute.fail_on_ansi_mode', False)
import pandas as pd
import numpy as np

# Read in the .csv file
nfl_df=ps.read_csv("data/weekly_nfl_data.csv")
```

```python
# Look at the first 5 rows
nfl_df.head(5)
```

- Once I load data, I like to look at column names and dtypes. Here I look at the columns in the NFL data so I can make meaningful decisions about what stats are of interest.

```python
# look at columns and datatypes in the nfl data

nfl_df.columns
#nfl_df.dtypes
```

- Below are quarterback statistics for the regular season from seasons 2005 to 2023. 


```python
nfl_seasons_df = nfl_df[(nfl_df["season"] >= 2005) & (nfl_df["season"]<=2023) 
                        & (nfl_df["position"] == "QB")
                        & (nfl_df["season_type"] == "REG")]

nfl_seasons_df.head(5)                          
```

 - I then looked at just the columns that include the player_display_name and some key statistics relevant to quarterbacks: season, week, completions, attempts, passing_yards, passing_tds, and interceptions. Based on the slack chat I removed week.

```python
nfl_brief_df = nfl_seasons_df[["player_display_name", "season", "completions",
                               "attempts", "passing_yards", "passing_tds", "interceptions"]]

nfl_brief_df.head(5)
```

- For each player_display_name and season combination, I found the sum and mean of each of the statistical quantities  included above.


```python
qb_stats = nfl_brief_df.groupby(["player_display_name", "season"]).agg(
    tot_completions = ("completions", "sum"),
    avg_completions = ("completions", "mean"),
    tot_attempts = ("attempts","sum"),
    avg_attempts = ("attempts", "mean"),
    tot_passing_yards = ("passing_yards", "sum"),
    avg_passing_yards = ("passing_yards", "mean"),
    tot_passing_tds = ("passing_tds", "sum"),
    avg_passing_tds = ("passing_tds", "mean"),
    tot_interceptions = ("interceptions", "sum"),
    avg_interceptions = ("interceptions", "mean"))

qb_stats.head()
```

- Here I've created two new variables (by season/player combination):

    completion_percentage = (sum of completions)/(sum of attempts)

    td_int_ratio = (sum passing tds)/(sum interceptions)
    
    **Because this is pandas-on-Spark, we need to watch out for dividing by zero.**
    In Spark SQL dividing by zero returns NaN. Because this is pandas-on-Spark, this throws an error instead. This is because ANSI mode is enabled as we were warned above.
    

```python
# creating completion percentage
qb_stats["completion_percentage"]=(qb_stats["tot_completions"]/ 
                                      qb_stats["tot_attempts"].where(qb_stats["tot_attempts"] != 0))

# creating td_int_ratio
qb_stats["td_int_ratio"]=(qb_stats["tot_passing_tds"]/ 
                              qb_stats["tot_interceptions"].where(qb_stats["tot_interceptions"] != 0))

qb_stats.head(20)
```

- Below I've filtered the rows to only include the player-season combinations where the attempts are >50.  Then I've sorted the rows in descending order by completion_percentage and reported the first 40 values!


```python
# filtering on attempts >50 
qb_stats50=qb_stats[qb_stats["tot_attempts"]>50]

#qb_stats50.head(10)

#and then sorting by completion percentage in descending order
qb_best=qb_stats50.sort_values("completion_percentage", ascending=False)

qb_best.head(40)

```


- Below I've filtered the rows to only include the player-season combinations where the attempts are >50. Then I've sorted the rows in descending order by td_int_ratio and reported the first 40 values. I had to drop the NaN rows because they showed up at the top otherwise.

```python
# sorting by best td-to-interceptions ratio in descending order
#had to drop na values because they were at the top which wasn't intended.
qb_best2=qb_stats50.sort_values("td_int_ratio", ascending=False).dropna(subset=["td_int_ratio"])

qb_best2.head(40)
```

### Spark SQL

In this section I use Spark SQL commands to look at the NFL data.  I start by reading in the file.

```python
# Read in the .csv file with pySpark
nfl_sql=spark.read.csv("data/weekly_nfl_data.csv",header=True, inferSchema=True)
```

- Here are the first 5 rows

```python
# looking at first 5 rows
nfl_sql.show(5)
```

- Here are all the column names.

```python
# view the columns in nfl data
nfl_sql.columns
```

- Below are quarterback statistics for the regular season from seasons 2005 to 2023. I'm only including the following columns (excluded week):

    player_display_name, season, completions, attempts, passing_yards, passing_tds, and interceptions


```python
#filtering on 2005-2023 regular season qb's
nfl_qb_sql = nfl_sql.filter(
    (nfl_sql["season"] >= 2005) & 
    (nfl_sql["season"] <= 2023) & 
    (nfl_sql["position"] == "QB") & 
    (nfl_sql["season_type"] == "REG")).select(
    "player_display_name", "season", "completions", 
    "attempts", "passing_yards", "passing_tds", "interceptions")
   

nfl_qb_sql.show(5)
```

- For each player_display_name and season combination, I found the sum and mean of each of the statistical quantities included above. I had to round the mean because it saved many decimal places.

```python
# aggregating stats
nfl_qb_agg= nfl_qb_sql.groupBy(["player_display_name", "season"]).agg(
     F.sum("completions"),F.round(F.mean("completions"), 2),
     F.sum("attempts"),F.round(F.mean("attempts"), 2),
     F.sum("passing_yards"),F.round(F.mean("passing_yards"), 2),
     F.sum("passing_tds"),F.round(F.mean("passing_tds"),2),
     F.sum("interceptions"),F.round(F.mean("interceptions"),2))
     
nfl_qb_agg.show(5)
```

- My column names have gotten long so I'm going to rename them.

```python
# renaming column summaries. 

# desired column names
c = ["player", "season", "sum_completions", "avg_completions", "sum_attempts", "avg_attempts", "sum_passing_yds", 
     "avg_passing_yds", "sum_passing_tds", "avg_passing_tds","sum_interceptions", "avg_interceptions"]

# loop through columns to rename
nfl_qb_stats=nfl_qb_agg
for i, col in enumerate(nfl_qb_agg.columns):
    nfl_qb_stats = nfl_qb_stats.withColumnRenamed(nfl_qb_stats.columns[i], c[i])

nfl_qb_stats.show(10)
```

- Now I'm creating two new variables (by season/player combination):

        completion_percentage = (sum of completions)/(sum of attempts)
    
        td_int_ratio = (sum passing tds)/(sum interceptions)


```python
#adding a completion percentage column
nfl_qb_stats1 = nfl_qb_stats.withColumn("completion_percentage",F.round(F.col("sum_completions")/F.col("sum_attempts"), 4))

#adding a touchdown to interception ratio
nfl_qb_stats2=nfl_qb_stats1.withColumn("td_int_ratio", F.round(F.col("sum_passing_tds")/F.col("sum_interceptions"),4))

nfl_qb_stats2.show(10)
```

**It's important to note that we did not  need to worry about dividing by 0 in these cases.  Spark SQL automatically assigns those values "NULL".**

- Below I've filtered the rows to only include the player-season combinations where the attempts are >50. Then I've sorted the rows in descending order by completion_percentage and reported the first 40 values.

```python
# filtering on attempts >50 and sorting by completion percentage in descending order
nfl_qb_best = nfl_qb_stats2.filter(F.col("sum_attempts") > 50).orderBy("completion_percentage", ascending=False)

nfl_qb_best.show(40)
```

- Below I've used the dataframe where the player-season combinations where the attempts are >50. Then I've sorted the rows in descending order by td_int_ratio and reported the first 40 values. **I did not have to drop NaN values like in pandas-on-Spark.**

```python
# sorting by td_int_ratio percentage in descending order
nfl_qb_best2 = nfl_qb_best.orderBy("td_int_ratio", ascending=False)

nfl_qb_best2.show(40)
```
